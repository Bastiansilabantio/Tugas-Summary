```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data penduduk
data = {
    'Provinsi': ['Jawa Barat', 'Jawa Timur', 'Jawa Tengah', 'Sumatera Utara', 'Sulawesi Selatan'],
    'Populasi': [50, 40, 36, 15, 9],  # Dalam juta
    'Luas Wilayah': [35377, 47800, 32800, 72981, 46717],  # Km²
    'Kepadatan': [1413, 837, 1098, 206, 193]  # Jiwa per km²
}

# Membuat DataFrame
df = pd.DataFrame(data)

# Set style and color palette
sns.set(style="whitegrid")
palette = sns.color_palette("Set2")

# Membuat subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Bar Plot untuk Populasi
sns.barplot(x='Provinsi', y='Populasi', data=df, palette=palette, ax=axes[0, 0])
axes[0, 0].set_title('Populasi Penduduk Berdasarkan Provinsi (Juta Jiwa)')
axes[0, 0].set_xlabel('Provinsi')
axes[0, 0].set_ylabel('Populasi (Juta Jiwa)')
for i, (prov, pop) in enumerate(zip(df['Provinsi'], df['Populasi'])):
    axes[0, 0].text(i, pop + 0.5, f'{pop}', ha='center')

# Plot 2: Scatter Plot Luas Wilayah vs Populasi
sns.scatterplot(x='Luas Wilayah', y='Populasi', hue='Provinsi', data=df, s=200, palette=palette, ax=axes[0, 1])
axes[0, 1].set_title('Populasi vs Luas Wilayah')
axes[0, 1].set_xlabel('Luas Wilayah (Km²)')
axes[0, 1].set_ylabel('Populasi (Juta Jiwa)')
for i, (luas, pop) in enumerate(zip(df['Luas Wilayah'], df['Populasi'])):
    axes[0, 1].text(luas, pop + 1, f'{pop} Juta', ha='right')

# Plot 3: Heatmap Kepadatan Penduduk
df_pivot = df.pivot_table(values='Kepadatan', index='Provinsi')
sns.heatmap(df_pivot, annot=True, cmap='YlOrRd', linewidths=.5, ax=axes[1, 0])
axes[1, 0].set_title('Kepadatan Penduduk per Km² Berdasarkan Provinsi')

# Plot 4: Bar Plot untuk Luas Wilayah
sns.barplot(x='Provinsi', y='Luas Wilayah', data=df, palette=palette, ax=axes[1, 1])
axes[1, 1].set_title('Luas Wilayah Berdasarkan Provinsi (Km²)')
axes[1, 1].set_xlabel('Provinsi')
axes[1, 1].set_ylabel('Luas Wilayah (Km²)')
for i, (prov, luas) in enumerate(zip(df['Provinsi'], df['Luas Wilayah'])):
    axes[1, 1].text(i, luas + 1000, f'{luas}', ha='center')

# Menyesuaikan layout
plt.tight_layout()
plt.show()
```
